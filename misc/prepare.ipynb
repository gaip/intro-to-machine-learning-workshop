# Prepare
This is the step where we prepare and wrangle the data. That means we're getting it ready to explore in the next step.

## Some things to consider and do:
- Set (or reset) the index 
- Are there columns that should hold derived data? For example, price * sales is revenue

```python
import pandas as pd
```

```python
df = pd.read_csv("lemonade.csv")

print(df.info())
```

```python
# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])


# Since the date column is unique, let's set the date column to be the index
df = df.set_index(df['Date']) 
df = df.drop("Date", 1) # drop the old Date column

df.head()
```

```python
# Let's create some derived columns
# Revenue is super useful, so let's do revenue!
df["Revenue"] = df.Price * df.Sales
```
