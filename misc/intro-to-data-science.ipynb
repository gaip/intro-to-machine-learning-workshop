# Intro to Data Science Workshop
In a nutshell, data science is the interdisciplinary approach of using names and numbers to determine the story the data can tell us. Our methods include programming, visualization, statistics, and machine learning.


## Data Science Workflow
1. Plan
2. Acquire
3. Prepare
4. Explore
5. Model
6. Deliver/Present

This workshop will touch on each piece of this pipeline, at a high level. 

## Plan
- What questions do you want to ask of your data?
- What's the business case for what we're looking at?
- What do we hope to gain by doing this analysis?

Remember the 5 kinds of questions that data science can answer:
1. How many/how much?
2. Which category does this thing belong to? Is this A or B or C?
3. What groupings or organization does the data already have?
4. Is this weird? 
5. What should we do next?
